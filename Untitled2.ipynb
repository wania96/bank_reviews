```python
import inflect

def remove_numbers(text):
    # Create an instance of the inflect engine
    p = inflect.engine()

    # Split the text into individual words
    words = text.split()

    # Loop through each word in the list
    for i in range(len(words)):
        # Check if the word is a number
        if words[i].isdigit():
            # Convert the number to words using inflect
            words[i] = p.number_to_words(words[i])
    
    # Join the words back into a string
    result = ' '.join(words)
    
    return result
```

```python
df['Reviews_clean']=df['Reviews_clean'].apply(remove_numbers)
```

```python
def remove_punct(text):
    text_nonpunc="".join([char for char in text if char not in string.punctuation])
    return text_nonpunc

df['Reviews_clean']=df['Reviews_clean'].apply(remove_punct)
```

```python
## Stopwords are unnecessary words like was, each etc.
## Tokenization means breaking text into individual words

import re

def tokenize(text):
    tokens = re.split('\W+', text)
    return tokens

df['Reviews_clean'] = df['Reviews_clean'].apply(lambda x: tokenize(x))

df.head()
```

```python
# NLTK - Natural Language Tool Kit is library to perform analysis on text 
# Corpus means bag of words 
#stopword = nltk.corpus.stopwords.words('english')

from nltk.corpus import stopwords
nltk.download('stopwords')
stop = stopwords.words('english')
```

```python
def remove_stopwords(tokenized_list):
    filtered_text=[word for word in tokenized_list if word not in stop]
    return filtered_text
```

```python
df['Reviews_clean']=df['Reviews_clean'].apply(lambda x:remove_stopwords(x))
```

```python
## Lemmatization is bring altogether all simmilar kind of words and create a single word out
import nltk
wn=nltk.stem.WordNetLemmatizer()

def lemmatizing(text):
    lemma=[wn.lemmatize(word) for word in text]
    return lemma
```

```python
df['Reviews_clean']=df['Reviews_clean'].apply(lemmatizing)
```

```python
word_count={}
for sentence in df['Reviews_clean']:
    for word in sentence:
        if word not in word_count:
            word_count[word]=1
        else:
            word_count[word]+=1  
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
nltk.download('vader_lexicon')
```

```python
analyser=SentimentIntensityAnalyzer()

def sentiment_pos(sentence):
    sent=analyser.polarity_scores(sentence)
    return sent['pos']

def sentiment_neg(sentence):
    sent=analyser.polarity_scores(sentence)
    return sent['neg']

def sentiment_neu(sentence):
    sent=analyser.polarity_scores(sentence)
    return sent['neu']

def sentiment_comp(sentence):
    sent=analyser.polarity_scores(sentence)
    return sent['comp']
```

```python
df_new['positive']=df_new['word'].apply(sentiment_pos)
df_new['negative']=df_new['word'].apply(sentiment_neg)
```

```python
from sklearn.cluster import KMeans
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
problems=df.loc[df.negative>0]['Reviews']
problems.head()
```

```python
vectorizer=TfidfVectorizer(stop_words='english')
X=vectorizer.fit_transform(problems)
```

```python
X
```

```python
## Now, implementing k-means algorithm to get the model so that we can fit into X
k=3
model=KMeans(n_clusters=3,init='k-means++',
    n_init=10,
    max_iter=300,
    tol=0.0001,
    verbose=0,
    random_state=None,
    copy_x=True,
    algorithm='auto')
```

```python
model.fit(X)
```

```python
## getting the centroids and features
centroids= model.cluster_centers_.argsort()[:, ::-1]
features=vectorizer.get_feature_names()
```

```python
for i in range(0,3):
    print('\n\nCluster %d:' % i)
    for ind in centroids[i, :10]:
        print('\t  - %s' % features[ind])
```

```python
### predicting the text sentence and tell into which cluster the sentence belongs to.
X=vectorizer.transform(df.Reviews)
df['cluster']=model.predict(X)
df.head()
```
